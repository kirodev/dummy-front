```python
import re
import PyPDF2
from PyPDF2 import PdfReader
import csv

# Define the PDF path
path = r'C:\Users\berga\Dropbox\Espacio familiar\Back-end\Data library\Judicial_documents\DE_Germany\Habich_Systematischer_Ueberblick_ueber_das_deutsche_FRAND_Fallrecht.pdf'

# Define patterns for court, date, and case
court_pattern = r'\s*LG\s*(München|Mannheim|Düsseldorf)'
date_pattern = r'\b(\d{1,2}\.\d{1,2}\.\d{4})\b'
case_pattern = r'\b(\d{1,2}\s*[a-z]?\s*O\s*\d{1,5}/\d{2})\b'

# Create an empty list to store extracted data
cases = []
last_court = None

# Open the PDF file
pdf_reader = PdfReader(path)
    
# Iterate over each page of the PDF
for page_num in range(len(pdf_reader.pages)):
    page = pdf_reader.pages[page_num]
    text = page.extract_text()
    
    # Find all potential matches for court, date, and case in the text
    courts = re.findall(court_pattern, text)
    dates = re.findall(date_pattern, text)
    cases_list = re.findall(case_pattern, text)
    
    # Determine the court for each case
    if courts:
        last_court = f'LG {courts[-1]}'  # Use the last found court
    else:
        if not last_court:
            continue  # Skip if there's no court information available
                      
    # Combine dates and cases, and add them to the cases list along with the determined court
    for date, case in zip(dates, cases_list):
        cases.append({'court': last_court, 'date': date, 'case': case})

# Write the extracted data to a CSV file
with open('cases.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=['court', 'date', 'case'])
    writer.writeheader()
    writer.writerows(cases)
```
