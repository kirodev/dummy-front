```python
import pandas as pd

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(r"C:\Users\berga\Dropbox\Espacio familiar\Back-end\Data library\Judicial_documents\DE_Germany\fights_raw.csv")

# Define functions to generate group level identifiers
def group_level_1(row):
    return f"{row['Landgericht']}-{row['Die Beklagte wird verurteilt zu unterlassen']}-{row['Entscheidung (Untrelassung)']}"

def group_level_2(row):
    return f"{group_level_1(row)}-{row['Urteil']}"

def group_level_3(row):
    return f"{group_level_1(row)}-{row['Date of complaint']}"

def group_level_4(row):
    return f"{group_level_1(row)}-{row['Pat application']}"

def group_level_5(row):
    return f"{group_level_2(row)}-{row['Pat application']}"

def group_level_6(row):
    return f"{group_level_3(row)}-{row['Pat application']}"

# Apply the functions to create the group level variables
df['Fight group level 1'] = df.apply(group_level_1, axis=1)
df['Fight group level 2'] = df.apply(group_level_2, axis=1)
df['Fight group level 3'] = df.apply(group_level_3, axis=1)
df['Fight group level 4'] = df.apply(group_level_4, axis=1)
df['Fight group level 5'] = df.apply(group_level_5, axis=1)
df['Fight group level 6'] = df.apply(group_level_6, axis=1)

# Create dictionaries to map unique combinations per level to numerical identifiers
group_mapping_1 = {group: idx for idx, group in enumerate(df['Fight group level 1'].unique(), start=101)}
group_mapping_2 = {group: idx for idx, group in enumerate(df['Fight group level 2'].unique(), start=201)}
group_mapping_3 = {group: idx for idx, group in enumerate(df['Fight group level 3'].unique(), start=301)}
group_mapping_4 = {group: idx for idx, group in enumerate(df['Fight group level 4'].unique(), start=401)}
group_mapping_5 = {group: idx for idx, group in enumerate(df['Fight group level 5'].unique(), start=501)}
group_mapping_6 = {group: idx for idx, group in enumerate(df['Fight group level 6'].unique(), start=601)}

# Create new columns for group identifiers based on each level's mapping
df['Group Identifier Level 1'] = df['Fight group level 1'].map(group_mapping_1)
df['Group Identifier Level 2'] = df['Fight group level 2'].map(group_mapping_2)
df['Group Identifier Level 3'] = df['Fight group level 3'].map(group_mapping_3)
df['Group Identifier Level 4'] = df['Fight group level 4'].map(group_mapping_4)
df['Group Identifier Level 5'] = df['Fight group level 5'].map(group_mapping_5)
df['Group Identifier Level 6'] = df['Fight group level 6'].map(group_mapping_6)

# Save the updated DataFrame to a new CSV file
df.to_csv(r"C:\Users\berga\Dropbox\Espacio familiar\Back-end\Data library\Judicial_documents\DE_Germany\fights_updated.csv", index=False)
```
